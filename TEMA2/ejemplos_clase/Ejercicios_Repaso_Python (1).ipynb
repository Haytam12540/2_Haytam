### Ejercicio 5 â€” Match (estructuras de selecciÃ³n mÃºltiples)
Pide un nÃºmero del 1 al 7 y muestra el nombre del dÃ­a de la semana.

```python
def dia_semana(n: int) -> str:
    match n:
        case 1:
            return 'Lunes'
        case 2:
            return 'Martes'
        case 3:
            return 'MiÃ©rcoles'
        case 4:
            return 'Jueves'
        case 5:
            return 'Viernes'
        case 6:
            return 'SÃ¡bado'
        case 7:
            return 'Domingo'
        case _:
            return 'NÃºmero invÃ¡lido'
```

```python
assert dia_semana(1) == 'Lunes'
assert dia_semana(7) == 'Domingo'
print('âœ… Ejercicio 5 correcto')
```

### Ejercicio 6 â€” Rango y suma de nÃºmeros pares
Suma todos los nÃºmeros pares entre 1 y 50.

```python
def suma_pares() -> int:
    suma = 0
    for num in range(2, 51, 2):
        suma += num
    return suma
```

```python
assert suma_pares() == sum(i for i in range(1,51) if i%2==0)
print('âœ… Ejercicio 6 correcto')
```

### Ejercicio 8 â€” Uso de break y continue
Pide una lista de nÃºmeros y devuelve la suma hasta encontrar el nÃºmero 0 (sin incluirlo).

```python
def suma_hasta_cero(numeros: list[int]) -> int:
    suma = 0
    for num in numeros:
        if num == 0:
            break
        suma += num
    return suma
```

```python
assert suma_hasta_cero([1,2,3,0,5]) == 6
assert suma_hasta_cero([5,5,0,10]) == 10
print('âœ… Ejercicio 8 correcto')
```

### Ejercicio 9 â€” Excepciones
Pide dos nÃºmeros y devuelve su divisiÃ³n. Si el divisor es 0, devuelve el texto 'Error: divisiÃ³n por cero'.

```python
def dividir(a: float, b: float) -> float|str:
    try:
        return a / b
    except ZeroDivisionError:
        return 'Error: divisiÃ³n por cero'
```

```python
assert dividir(10,2) == 5
assert dividir(8,0) == 'Error: divisiÃ³n por cero'
print('âœ… Ejercicio 9 correcto')
```

### Ejercicio 10 â€” Juego de adivinanza
Genera un nÃºmero secreto del 1 al 10. El jugador intenta adivinarlo. Devuelve 'Correcto' si acierta y 'Incorrecto' si no.

```python
def adivinar(num: int, secreto: int) -> str:
    return 'Correcto' if num == secreto else 'Incorrecto'
```

```python
assert adivinar(5,5) == 'Correcto'
assert adivinar(3,7) == 'Incorrecto'
print('âœ… Ejercicio 10 correcto')
```
